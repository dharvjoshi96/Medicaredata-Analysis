```python
#CMS states that higher HCC scores means there is higher risk of chronic diseases while lower hcc indicates healthy population. 
#It will be interesting to see what is the average hcc score for different states and facilities.

# Max AND Min HCC scores for different states


query3 = """SELECT
  state, AVG(average_hcc_score) as avg_hcc_score, Max(average_hcc_score) as max_hcc_score, 
  Min(average_hcc_score) as min_hcc_score, COUNT(average_hcc_score) as total_facilities
FROM
  `bigquery-public-data.cms_medicare.nursing_facilities_2014`
  GROUP BY 
  state
ORDER BY
  avg_hcc_score DESC;
        """
```

```python
query_job2 = client.query(query3)
HCC_Scores = query_job2.to_dataframe()
```

```python
HCC_Scores.head(10)
```

```python
HCC_Scores.iplot(kind='bar', x='state' )
```

```python
#Finding States / Facilities with poor average length of stay days

query5 =  """SELECT
  state, AVG(average_length_of_stays_days) as average_length_of_stays_days, AVG(total_stays) as total_stays
FROM
  `bigquery-public-data.cms_medicare.nursing_facilities_2014`
  GROUP BY 
  state
ORDER BY
  average_length_of_stays_days DESC;"""
```

```python
#Explore the Total Claims, Day Supply and Drug Costs



query6 = """SELECT
  nppes_provider_state AS state,
  ROUND(SUM(total_claim_count) / 1e6) AS total_claim_count_millions,
  ROUND(SUM(total_day_supply) / 1e6) AS total_day_supply_millions,
  ROUND(SUM(total_drug_cost) / 1e6) AS total_drug_cost_millions
FROM
  `bigquery-public-data.cms_medicare.part_d_prescriber_2014`
GROUP BY
  state
ORDER BY
  total_claim_count_millions DESC;
"""
```

```python
query_job6 = client.query(query6)
Claims_2014 = query_job6.to_dataframe()
```

```python
Claims_2014.head(10)
```

```python
import plotly.express as px
fig = px.scatter( Claims_2014, x="total_claim_count_millions", y="total_day_supply_millions",
 hover_name="state" )
```

```python
query7 = """SELECT
  generic_name AS drug_name,
  ROUND(SUM(total_claim_count) / 1e6) AS total_claim_count_millions,
  ROUND(SUM(total_day_supply) / 1e6) AS total_day_supply_millions,
  ROUND(SUM(total_drug_cost) / 1e6) AS total_drug_cost_millions
FROM
  `bigquery-public-data.cms_medicare.part_d_prescriber_2014`
GROUP BY
  drug_name
ORDER BY
  total_claim_count_millions DESC;
 """
```

```python
query_job7 = client.query(query7)
Drugs_2014 = query_job7.to_dataframe()
```

```python
Drugs_2014.head()
```

```python
Drugs_2014.iplot(kind='barh', x = 'drug_name')
```

```python
#Nursing Specialities associated with highest claims

query8 = """SELECT
  specialty_description AS specialty_description,
  ROUND(SUM(total_claim_count) / 1e6) AS total_claim_count_millions,
  ROUND(SUM(total_day_supply) / 1e6) AS total_day_supply_millions,
  ROUND(SUM(total_drug_cost) / 1e6) AS total_drug_cost_millions
FROM
  `bigquery-public-data.cms_medicare.part_d_prescriber_2014`
GROUP BY
  specialty_description
ORDER BY
  total_claim_count_millions DESC;
"""
query_job8 = client.query(query8)
Diff_Claims_2014 = query_job7.to_dataframe()
```

```python
Diff_Claim_2014_head = Diff_Claims_2014.head(20)

import plotly.express as px
fig = px.pie(Diff_Claim_2014_head, values='total_claim_count_millions', names='drug_name',
title='Nursing Specialities associated with highest claims')
fig.show()

```
